13.  Implement K-Means clustering on Iris.csv dataset. Determine the number of clusters using the elbow method. Dataset Link: https://www.kaggle.com/datasets/uciml/iris

Step 1: Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

Step 2: Load Dataset

```python
df = pd.read_csv("Iris.csv")
print(df.head())
```

Step 3: Data Preprocessing
1. Drop unnecessary column

```python
df.drop("Id", axis=1, inplace=True)
```

2. Separate features

```python
X = df.drop("Species", axis=1)
```

3. Feature Scaling

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Step 4: Elbow Method (Find Optimal K)

```python
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
```

Step 5: Plot Elbow Graph

```python
plt.plot(range(1, 11), wcss, marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("WCSS")
plt.show()
```

Step 6: Apply K-Means

```python
kmeans = KMeans(n_clusters=3, random_state=42)
y_kmeans = kmeans.fit_predict(X_scaled)
```

Step 7: Visualize Clusters

(Using 2 features for plotting)

```python
plt.scatter(X_scaled[:, 2], X_scaled[:, 3], c=y_kmeans, cmap='viridis')
plt.title("K-Means Clustering (Iris Dataset)")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.show()
```

Step 8: Cluster Centers

```python
print("Cluster Centers:\n", kmeans.cluster_centers_)
```
